1. Для нахождения примитивного элемента напишем условие: $\alpha^{q-1}=1$ и $\alpha^k != 1$ при $0 < k < q-1$. Для этого напишем программу на Python, чтобы их найти. Так как нам надо, чтобы выполнялось условие $q^m - 1 = n$. Из условия $n=12$, $\delta=5$. Единственный возможный вариант: $q = 13$, $m = 1$. После работы программы видно, что таких элементов может быть много, но я возьму наименьший, чтобы было удобно считать. Следовательно $\alpha = 2$.

```python
new = []
for i in range(2, 20):
    q = [i ** j % 13 for j in range(1, 13)]
    if q.count(1) == 1:
        new.append(i)
print(new)
```

4. Для того чтобы построить порождающий многочлен $g(x)$ нам нужно знать минимальные многочлены $m_0(x), m_1(x), m_2(x), m_3(x)$, соответствующие элементам $\alpha, \alpha^2, \alpha^3, \alpha^4$ (так как $\delta=5$). Минимальные многочлены:
   $m_0(x) = x - \alpha$
   $m_1(x) = x - \alpha^2$
   $m_2(x) = x - \alpha^3$
   $m_3(x) = x - \alpha^4$
   Тогда $g(x) = (x - \alpha)(x - \alpha^2)(x - \alpha^3)(x - \alpha^4)$. В данное уравнение вместо $\alpha^i$ можно подставить значения из 2 пункта. Получим $g(x) = (x - 2)(x - 4)(x - 8)(x - 3) = x^4 + 9x^3 + 7x^2 + 2x + 10$. Вычисления проводятся по модулю $13$. Также можно считать в общем виде, для это можно написать такой код, который строит таблицы для поля Галуа.


```python
import galois


field = galois.GF(13 ** 1, repr='power')
x = field([i for i in range(1, 13)])

print(x.arithmetic_table('+'))
print(x.arithmetic_table('*'))
print(x.arithmetic_table('-'))
```
