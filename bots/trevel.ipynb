## ТРЕВЕЛГИД
Данный чат бот помогает путешественникам выбрать авиа или жд билеты, также предлагает построить маршрут на машине, отправляя ссылку с маршрутом в "run" с помощью библиотеки 'urllib.parse'. Ссылку на документацию я прикрепила в коде сразу после импорта.

```python
import urllib.parse

"""
https://digitology.tech/docs/python_3/library/urllib.parse.html
ссылка на документацию данной библиотеки
"""

# приветствие бота
print("Здравствуйте! Добро пожаловать, я - ТревелГид")
# поздоровайтесь с ботом
input()
print('Как вас зовут?')
# вводим в переменную 'name' имя, которое будет использоваться дальше
# с помощью .title() делаем так, чтобы первая буква была заглавной, а остальные - прописными
name = input('Введите ваше имя:\n').title()
# с помощью f строки, обращаемся к переменной 'name'
print(f'Приятно познакомиться, {name}')
print(f'{name}, как ваше настроение? Готовы к общению?')
# пользователь отвечает произвольное сообщение
input()

# Список доступных команд
commands = {1: 'Хочу купить авиабилеты', 2: 'Хочу купить билеты на поезд',
            3: 'Проложи путь на машине', 4: 'Я передумал к вам обращаться'}

# Список доступных городов
cities = {1: 'Москва', 2: 'Санкт-Петербург', 3: 'Екатеринбург',
          4: 'Сочи', 5: 'Краснодар', 6: 'Новосибирск'}

# Список авиакомпаний
companies = {1: 'Аэрофлот', 2: 'S7', 3: 'Red Wings', 4: 'Победа', 5: 'Utair'}

# Список аэропортов
airports = {1: 'Домодедово', 2: 'Шереметьево', 3: 'Внуково'}

# Список железнодорожных станций
railway_stations = {1: 'Белорусский ', 2: 'Казанский', 3: 'Киевский',
                    4: 'Курский', 5: 'Ленинградский', 6: 'Павелецкий'}

# Список доступных дат
dates = {1: '01.01.2024', 2: '02.02.2024',
         3: '03.03.2024', 4: '04.04.2024', 5: '05.05.2024'}

# Список типов авиабилетов
type_of_air_tickets = {1: 'Бизнес', 2: 'Эконом'}

# Список типов железнодорожных билетов
type_of_train_tickets = {1: 'Сидячий', 2: 'Плацкарт',
                         3: 'Купе', 4: 'СВ', 5: 'Люкс'}

print(f'{name}, прилагаю Вам список команд, по которым я могу Вас сориентировать')
# с помощью цикла выводим ключи и значения словаря, чтобы пользователь мог выбрать нужную команду
for keys, values in commands.items():
    print(f'{keys} - {values}')

# запускаем бесконечный цикл 'while'
while True:
    # пользователь вводит число, чтобы выбрать нужную команду
    command = int(input(f'{name}, введите одну из предложенных цифр:\n'))
    # проверяем правильную ли цифру ввёл пользователь
    if command not in commands.keys():
        print(f'{name}, я Вас не понимаю, введите сообщение повторно, пожалуйста')
    # дополнительное условие, если пользователь выбрал авиабилеты
    elif commands[command] == 'Хочу купить авиабилеты':
        print(f'{name}, какого класса билеты вы бы хотели купить?')
        # с помощью цикла 'for' выводим ключи и значения словаря 'type_of_air_tickets'
        for keys, values in type_of_air_tickets.items():
            print(f'{keys} - {values}')
        # запускаем бесконечный цикл 'while'
        while True:
            type = int(input())
            # проверяем правильно ли ввёл пользователь цифру,
            # если нет, его переносит в 'else' и он вводит её до тех пор,
            # пока она не будет находиться в ключах данного словаря
            if type in type_of_air_tickets.keys():
                print(f'Хорошо, {name}, я запомнил, что вам нужны билеты {type} класса')
                # выходим из внутреннего цикла, если пользователь выбрал правильное значение
                break
            else:
                print(f'{name}, выберите одну цифру из представленных выше')

        print(f'{name}, теперь давайте определимся с авиакомпанией, которой вы хотите лететь')
        # с помощью цикла 'for' выводим ключи и значения словаря 'companies'
        for keys, values in companies.items():
            print(f'{keys} - {values}')
        # запускаем бесконечный цикл 'while'
        while True:
            company = int(input())
            # проверяем правильно ли ввёл пользователь цифру,
            # если нет, его переносит в 'else' и он вводит её до тех пор,
            # пока она не будет находиться в ключах данного словаря
            if company in companies.keys():
                print(f'Отличный выбор, {name}! Вы точно полетите данной авиакомпанией: {companies[company]}')
                # выходим из внутреннего цикла, если пользователь выбрал правильное значение
                break
            else:
                print(f'{name}, выберите одну цифру из представленных выше')

        print(f'{name}, какой аэропорт Вам больше всего симпатизирует?')
        # с помощью цикла 'for' выводим ключи и значения словаря 'airports'
        for keys, values in airports.items():
            print(f'{keys} - {values}')
        # запускаем бесконечный цикл 'while'
        while True:
            airport = int(input())
            # проверяем правильно ли ввёл пользователь цифру,
            # если нет, его переносит в 'else' и он вводит её до тех пор,
            # пока она не будет находиться в ключах данного словаря
            if airport in airports.keys():
                print(f'{name}, буду знать, что вы любите {airports[airport]}')
                # выходим из внутреннего цикла, если пользователь выбрал правильное значение
                break
            else:
                print(f'{name}, выберите одну цифру из представленных выше')

        print(f'{name}, в какой город полетим?')
        # с помощью цикла 'for' выводим ключи и значения словаря 'cities'
        for keys, values in cities.items():
            print(f'{keys} - {values}')
        # запускаем бесконечный цикл 'while'
        while True:
            city = int(input())
            if city in cities.keys():
                print(f'{name}, держим курс на {cities[city]}')
                # выходим из внутреннего цикла, если пользователь выбрал правильное значение
                break
            else:
                print(f'{name}, выберите одну цифру из представленных выше')

        print(f'{name}, какой день Вам больше нравится?')
        # с помощью цикла 'for' выводим ключи и значения словаря 'dates'
        for keys, values in dates.items():
            print(f'{keys} - {values}')
        # запускаем бесконечный цикл 'while'
        while True:
            date = int(input())
            # проверяем правильно ли ввёл пользователь цифру,
            # если нет, его переносит в 'else' и он вводит её до тех пор,
            # пока она не будет находиться в ключах данного словаря
            if date in dates.keys():
                print(f'{name}, отличная дата! {dates[date]}')
                # выходим из внутреннего цикла, если пользователь выбрал правильное значение
                break
            else:
                print(f'{name}, выберите одну цифру из представленных выше')

        print(f'{name}, вы летите: {dates[date]}\n'
              f'Авиакомпанией: {companies[company]}\n'
              f'Из аэропорта: {airports[airport]}\n'
              f'{type_of_air_tickets[type]} классом\n'
              f'В город: {cities[city]}')
        print('Спасибо, что пользуетесь нашим тревелгидом! Посоветуете ли вы его друзьям?')
        print('Да/нет')
        input()
        print(f'{name}, спасибо за ваш ответ, он очень важен для нас!')
        break


    elif commands[command] == 'Хочу купить билеты на поезд':

        print(f'{name}, в каком типе вагона вы хотите ехать?')
        # с помощью цикла 'for' выводим ключи и значения словаря 'type_of_train_tickets'
        for keys, values in type_of_train_tickets.items():
            print(f'{keys} - {values}')
        # запускаем бесконечный цикл 'while'
        while True:
            type = int(input())
            # проверяем правильно ли ввёл пользователь цифру,
            # если нет, его переносит в 'else' и он вводит её до тех пор,
            # пока она не будет находиться в ключах данного словаря
            if type == 1 or type == 2 or type == 3 or type == 4 or type == 5:
                print(f'{name}, бронируем данный вид билетов! {type_of_train_tickets[type]}')
                # выходим из внутреннего цикла, если пользователь выбрал правильное значение
                break
            else:
                print(f'{name}, выберите одну цифру из представленных выше')

        print(f'{name}, в какой город поедем?')
        # с помощью цикла 'for' выводим ключи и значения словаря 'cities'
        for keys, values in cities.items():
            print(f'{keys} - {values}')
        # запускаем бесконечный цикл 'while'
        while True:
            city = int(input())
            # проверяем правильно ли ввёл пользователь цифру,
            # если нет, его переносит в 'else' и он вводит её до тех пор,
            # пока она не будет находиться в ключах данного словаря
            if city in cities.keys():
                print(f'{name}, держим курс на {cities[city]}')
                # выходим из внутреннего цикла, если пользователь выбрал правильное значение
                break
            else:
                print(f'{name}, выберите одну цифру из представленных выше')

        print(f'{name}, с какого вокзала?')
        # с помощью цикла 'for' выводим ключи и значения словаря 'railway_stations'
        for keys, values in railway_stations.items():
            print(f'{keys} - {values}')
        # запускаем бесконечный цикл 'while'
        while True:
            railway_station = int(input())
            # проверяем правильно ли ввёл пользователь цифру,
            # если нет, его переносит в 'else' и он вводит её до тех пор,
            # пока она не будет находиться в ключах данного словаря
            if railway_station in railway_stations.keys():
                print(f'{name}, держим курс на {railway_stations[railway_station]}')
                # выходим из внутреннего цикла, если пользователь выбрал правильное значение
                break
            else:
                print(f'{name}, выберите одну цифру из представленных выше')

        print(f'{name}, какую дату предпочитаете?')
        # с помощью цикла 'for' выводим ключи и значения словаря 'dates'
        for keys, values in dates.items():
            print(f'{keys} - {values}')
        # запускаем бесконечный цикл 'while'
        while True:
            date = int(input())
            # проверяем правильно ли ввёл пользователь цифру,
            # если нет, его переносит в 'else' и он вводит её до тех пор,
            # пока она не будет находиться в ключах данного словаря
            if date in dates.keys():
                print(f'{name}, отличная дата! {dates[date]}')
                # выходим из внутреннего цикла, если пользователь выбрал правильное значение
                break
            else:
                print(f'{name}, выберите одну цифру из представленных выше')

        print(f'{name}, Вы едете: {dates[date]}\n'
              f'Вокзал: {railway_stations[railway_station]}\n'
              f'Вагон типа: {type_of_train_tickets[type]}n'
              f'В город: {cities[city]}')

        print('Спасибо, что пользуетесь нашим тревелгидом! Посоветуете ли вы его друзьям?')
        print('Да/нет')
        input()
        print(f'{name}, спасибо за ваш ответ, он очень важен для нас!')
        break

    elif commands[command] == 'Проложи путь на машине':
        start = input(f'{name}, введите город отправления: ')
        end = input(f'{name}, введите город назначения: ')

        # Преобразовываем города в URL-совместимую строку
        start_url = urllib.parse.quote(start)
        end_url = urllib.parse.quote(end)

        # Формируем ссылку на Google Maps с использованием введенных городов
        url = f"https://www.google.com/maps/dir/?api=1&origin={start_url}&destination={end_url}"
        # выводим данный маршрут
        print(f'{name}, вот ваш маршрут: {url}')

    else:
        print(f'{name}, до скорых встреч! Надеемся, что вы вернётесь и будете путешествовать с нами')
        break
```
